```python
def last3_close_gt_open(open_series: pd.Series, close_series: pd.Series) -> bool:
    """
    Returns True if the last 3 days' closes are greater than opens.
    """
    diff = close_series - open_series
    last3 = diff.tail(3)
    check_diff = (last3 > 0).all()
    return check_diff
```

```python
def last3_gt_prev(series: pd.Series) -> bool:
    last3 = series.tail(3)  # need 4 values to compare 3 consecutive increases
    return (last3.diff().tail(2) > 0).all()
```

```python
c_list = []
for item in common:
    # print(item)
    close_series = df.T["Close"][item]
    open_series = df.T["Open"][item]

    check1 = last3_close_gt_open(open_series, close_series)
    check2 = last3_gt_prev(open_series)
    check3 = last3_gt_prev(close_series)

    if check1 and check2 and check3:
        c_list.append(item)

print(len(c_list))
```

```python
def stock_health_check(data: dict) -> dict:
    results = {}

    # --- P/E Ratios ---
    results["Trailing P/E"] = data.get("trailingPE")
    results["Forward P/E"] = data.get("forwardPE")

    # --- Cash Flow per Share ---
    ocf = data.get("operatingCashflow")
    shares = data.get("sharesOutstanding")
    if ocf and shares:
        results["Cash Flow per Share"] = ocf / shares
    else:
        results["Cash Flow per Share"] = None

    # --- EPS ---
    results["EPS (TTM)"] = data.get("trailingEps")

    # --- Liquidity ---
    results["Current Ratio"] = data.get("currentRatio")
    results["Quick Ratio"] = data.get("quickRatio")

    # --- Returns ---
    results["Return on Assets"] = data.get("returnOnAssets")
    results["Return on Equity"] = data.get("returnOnEquity")

    # --- Dividend ---
    results["Dividend Yield %"] = data.get("dividendYield")
    results["Payout Ratio"] = data.get("payoutRatio")

    # --- Debt ---
    results["Debt-to-Equity"] = data.get("debtToEquity")

    # --- Basic Health Verdict ---
    verdict = []
    if results["Forward P/E"] and results["Forward P/E"] < 15:
        verdict.append("Cheap valuation")
    if results["Cash Flow per Share"] and results["EPS (TTM)"] and results["Cash Flow per Share"] > results["EPS (TTM)"]:
        verdict.append("Strong cash flow quality")
    if results["Current Ratio"] and results["Current Ratio"] > 1.2:
        verdict.append("Healthy liquidity")
    if results["Return on Equity"] and results["Return on Equity"] > 0.15:
        verdict.append("High efficiency (ROE)")
    if results["Dividend Yield %"] and results["Dividend Yield %"] > 0.05:
        verdict.append("Attractive dividend")

    results["Verdict"] = verdict
    return results
```

```python
def stock_profile(data: dict) -> dict:
    results = {}

    # --- Extract Key Values ---
    trailing_pe = data.get("trailingPE")
    forward_pe = data.get("forwardPE")
    eps = data.get("trailingEps")
    ocf = data.get("operatingCashflow")
    shares = data.get("sharesOutstanding")
    cash_per_share = data.get("totalCashPerShare")
    dividend_yield = data.get("dividendYield")
    payout_ratio = data.get("payoutRatio")
    current_ratio = data.get("currentRatio")
    quick_ratio = data.get("quickRatio")
    roe = data.get("returnOnEquity")
    roa = data.get("returnOnAssets")
    debt_to_equity = data.get("debtToEquity")
    revenue_growth = data.get("revenueGrowth")
    earnings_growth = data.get("earningsGrowth")
    profit_margin = data.get("profitMargins")

    # --- Derived Values ---
    if ocf and shares:
        cash_flow_per_share = ocf / shares
    else:
        cash_flow_per_share = None

    # Save key outputs
    results.update({
        "Trailing P/E": trailing_pe,
        "Forward P/E": forward_pe,
        "EPS (TTM)": eps,
        "Cash Flow per Share": cash_flow_per_share,
        "Cash per Share": cash_per_share,
        "Dividend Yield %": dividend_yield,
        "Payout Ratio": payout_ratio,
        "Current Ratio": current_ratio,
        "Quick Ratio": quick_ratio,
        "ROE": roe,
        "ROA": roa,
        "Debt-to-Equity": debt_to_equity,
        "Revenue Growth": revenue_growth,
        "Earnings Growth": earnings_growth,
        "Profit Margin": profit_margin
    })

    # --- Scoring ---
    verdict = []

    # Valuation
    if forward_pe and forward_pe < 15:
        verdict.append("Cheap valuation")
    elif trailing_pe and trailing_pe > 20:
        verdict.append("Expensive valuation")

    # Cash / Dividend
    if cash_flow_per_share and eps and cash_flow_per_share > eps:
        verdict.append("Strong cash flow quality")
    if cash_per_share and data.get("currentPrice") and cash_per_share > data["currentPrice"] * 0.4:
        verdict.append("High cash reserves")
    if dividend_yield and dividend_yield > 0.05:
        verdict.append("High dividend yield")
    elif dividend_yield and dividend_yield < 0.02:
        verdict.append("Low dividend")

    # Liquidity
    if current_ratio and current_ratio < 1.2:
        verdict.append("Weak liquidity")
    elif current_ratio and current_ratio > 1.5:
        verdict.append("Healthy liquidity")

    # Profitability
    if roe and roe > 0.15:
        verdict.append("High efficiency (ROE)")
    if roa and roa < 0.03:
        verdict.append("Low ROA")
    if profit_margin and profit_margin < 0.05:
        verdict.append("Thin margins")

    # Debt
    if debt_to_equity and debt_to_equity > 70:
        verdict.append("High debt load")

    # Growth
    if revenue_growth and revenue_growth > 0.15:
        verdict.append("Strong revenue growth")
    if earnings_growth and earnings_growth > 0.3:
        verdict.append("Explosive earnings growth")

    results["Verdict"] = verdict

    # --- Classification ---
    profile = "Balanced"
    if "Cheap valuation" in verdict and ("High dividend yield" in verdict or "Strong cash flow quality" in verdict):
        profile = "Cash Cow / Value"
    elif "Expensive valuation" in verdict and ("Strong revenue growth" in verdict or "Explosive earnings growth" in verdict):
        profile = "Growth Bet"
    elif "Weak liquidity" in verdict or "High debt load" in verdict or "Thin margins" in verdict:
        profile = "High Risk"

    results["Profile"] = profile
    return results
```

```python
def score_stock(data: dict) -> dict:
    # Extract metrics
    metrics = {
        "forward_pe": data.get("forwardPE"),
        "trailing_pe": data.get("trailingPE"),
        "eps": data.get("trailingEps"),
        "ocf": data.get("operatingCashflow"),
        "shares": data.get("sharesOutstanding"),
        "cash_per_share": data.get("totalCashPerShare"),
        "current_price": data.get("currentPrice"),
        "div_yield": data.get("dividendYield"),
        "payout": data.get("payoutRatio"),
        "curr_ratio": data.get("currentRatio"),
        "quick_ratio": data.get("quickRatio"),
        "roe": data.get("returnOnEquity"),
        "roa": data.get("returnOnAssets"),
        "debt_eq": data.get("debtToEquity"),
        "rev_growth": data.get("revenueGrowth"),
        "earnings_growth": data.get("earningsGrowth"),
        "profit_margin": data.get("profitMargins")
    }

    # Compute cash flow per share
    if metrics["ocf"] and metrics["shares"]:
        metrics["cfps"] = metrics["ocf"] / metrics["shares"]
    else:
        metrics["cfps"] = None

    verdict = []
    score = 50  # neutral starting point

    # Valuation adjustments
    if metrics["forward_pe"] and metrics["forward_pe"] < 15:
        verdict.append("Cheap valuation"); score += 10
    if metrics["trailing_pe"] and metrics["trailing_pe"] != "Infinity"  and metrics["trailing_pe"] > 20:
        verdict.append("Expensive valuation"); score -= 10

    # Cash and dividend
    if metrics["cfps"] and metrics["eps"] and metrics["cfps"] > metrics["eps"]:
        verdict.append("Strong cash flow"); score += 5
    if metrics["cash_per_share"] and metrics["current_price"] and metrics["cash_per_share"] > metrics["current_price"] * 0.4:
        verdict.append("High cash reserves"); score += 5
    if metrics["div_yield"] and metrics["div_yield"] > 0.05:
        verdict.append("High dividend yield"); score += 5
    elif metrics["div_yield"] and metrics["div_yield"] < 0.02:
        verdict.append("Low dividend"); score -= 5

    # Liquidity & debt
    if metrics["curr_ratio"] and metrics["curr_ratio"] < 1:
        verdict.append("Weak liquidity"); score -= 5
    elif metrics["curr_ratio"] and metrics["curr_ratio"] > 1.5:
        verdict.append("Healthy liquidity"); score += 5
    if metrics["debt_eq"] and metrics["debt_eq"] > 70:
        verdict.append("High debt load"); score -= 5

    # Profitability
    if metrics["roe"] and metrics["roe"] > 0.15:
        verdict.append("High efficiency (ROE)"); score += 5
    if metrics["roa"] and metrics["roa"] < 0.03:
        verdict.append("Low ROA"); score -= 5
    if metrics["profit_margin"] and metrics["profit_margin"] < 0.05:
        verdict.append("Thin margins"); score -= 5

    # Growth factors
    if metrics["rev_growth"] and metrics["rev_growth"] > 0.15:
        verdict.append("Strong revenue growth"); score += 5
    if metrics["earnings_growth"] and metrics["earnings_growth"] > 0.3:
        verdict.append("Explosive earnings growth"); score += 5

    # Bound the score
    score = max(0, min(100, score))

    # Profile classification
    profile = "Balanced"
    if "Cheap valuation" in verdict and ("High dividend yield" in verdict or "Strong cash flow" in verdict):
        profile = "Cash Cow / Value"
    elif "Expensive valuation" in verdict and ("Strong revenue growth" in verdict or "Explosive earnings growth" in verdict):
        profile = "Growth Bet"
    if "Weak liquidity" in verdict or "High debt load" in verdict or "Thin margins" in verdict:
        profile = "High Risk"

    return {"score": score, "verdict": verdict, "profile": profile}
```

```python
def check_trend(pct_series: pd.Series):
    # condition 1: 2 consecutive >= +1%
    cond1 = ((pct_series >= 1).astype(int)
             .rolling(2).sum() >= 2).any()

    # condition 2: 3 consecutive > 0
    cond2 = ((pct_series > 0).astype(int)
             .rolling(3).sum() >= 3).any()

    return cond1, cond2
```
